Krenimo od 1. primjera. Znaci prva dva reda predstavljaju produzenje ugovora(odatle i duplikati, pogledati FAQ). Produzenje ugovora se dogodilo prije PDZ(planiranog datuma zatvaranja) te je novi PDZ razumno vremenski udaljen od novog DO(datuma otvaranja). Isto to se dogada sa druga dva reda samo sto je netko produzio ugovor tocno na PDZ. Kod ovog primjera nije nista cudno. Pretpostavljamo da je to tako upravo zbog gore opisanih stvari.

Pogledajmo sada 2. primjer na kojem vidimo drugu situaciju. Partija je produzivana, ali tek nakon PDZ. U takvom slucaju, datumi duplikata(produzenih ugovora) podivljaju. Nekad je DO=PDZ, nekad DZ>PDZ. Nasa nekakva teorija je da su ti ugovori sa ludim odnosom DO-a i PDZ-a zapravo partije(ugovori) koji su sklopljeni samo radi neke bankarske formalnosti kada je korisnik vratio sve sto je trebao. 
Treba spomenuti i zadnja dva reda ciji prvi(najranije otvoreni) ugovor izgleda nemamo jer logicno je da bi on trebao postojati i da bi trebao imati smislen odnos DO i PDZ tj. da vrijedi DO<PDZ

Treba napomenuti da je ovo sve zakljuceno iz samog promatranja tablice 'duplikati' te bi bilo dobro ove navode i provjeriti:

```python
# fokusirat cemo se samo na sljedece stupce. UGOVORENI_IZNOS i VISINA_KAMATE nam mozda kazu nesto korisno takoder
duplikati = duplikati[['OZNAKA_PARTIJE', 'DATUM_OTVARANJA', 'PLANIRANI_DATUM_ZATVARANJA', 'UGOVORENI_IZNOS', 'VISINA_KAMATE']]
duplikati.set_index('OZNAKA_PARTIJE', inplace=True)
```

**NASA INTERNA OBJASNJENJA ZA CELIJU ISPOD:**
- ugovor je *normalan* ako je njegov DATUM_OTVARANJA **manji** od PLANIRANI_DATUM_ZATVARANJA
- ugovor je *nenormalan* ako je njegov DATUM_OTVARANJA **veci** od PLANIRANI_DATUM_ZATVARANJA
- tablica 'isti_ugovori' sadrzi sve redove neke oznake koja se pojavljuje u tablici 'duplikati'.
Svakom for iteracijom, tablica 'isti_ugovori' se mijenja u skladu sa izabranom oznakom

```python
normalan_prvi_ugovor  = pd.DataFrame()
prvi_ugovor_vracen_isti_dan = pd.DataFrame()
nenormalan_prvi_ugovor = pd.DataFrame()


# idi po svim duplikatima(ali samo po jednom od svakog duplikata!)
for ugovor in duplikati.index.drop_duplicates():
    # zelimo da nam na prvom mjestu bude ugovor koji je najranije otvoren pa stoga sortiramo prilikom po DATUM_OTVARANJA
    isti_ugovori = duplikati.loc[[ugovor]].sort_values(by='DATUM_OTVARANJA')
    # resetirat cemo indeks kako bi mogli lakse pristupati redovima tablice 'isti_ugovori'
    isti_ugovori.reset_index(inplace=True)
    
    # ovdje pronademo kojoj skupini pripada taj prvi ugovor te sve ostale ugovore
    # (koji imaju DATUM_OTVARANJA veci nego DATUM_OTVARANJA promatranog ugovora) spremimo u tablicu kojoj pripadaju
    if (isti_ugovori.loc[[0]]['PLANIRANI_DATUM_ZATVARANJA'] - isti_ugovori.loc[[0]]['DATUM_OTVARANJA'] > pd.Timedelta(days=0)).item():
        normalan_prvi_ugovor = normalan_prvi_ugovor.append(isti_ugovori.loc[1:])
                
    elif (isti_ugovori.loc[[0]]['PLANIRANI_DATUM_ZATVARANJA'] - isti_ugovori.loc[[0]]['DATUM_OTVARANJA'] == pd.Timedelta(days=0)).item():
        prvi_ugovor_vracen_isti_dan = prvi_ugovor_vracen_isti_dan.append(isti_ugovori.loc[1:])
        
    else:
        nenormalan_prvi_ugovor = nenormalan_prvi_ugovor.append(isti_ugovori.loc[1:])


        
normalan_prvi_ugovor.name = 'Ugovori ciji je prvi ugovor normalan'
prvi_ugovor_vracen_isti_dan.name = 'Ugovori ciji je prvi ugovor trebao biti vracen na dan posudivanja novaca'
nenormalan_prvi_ugovor.name = 'Ugovori ciji je prvi ugovor nenormalan'
```

```python
def analiziraj_datume(tablica):
    print(tablica.name + ': ')
    print('Sveukupno: ' + str(tablica.shape[0]))
    print('Koliko datuma ima DO > PDZ: ' + str(tablica[(tablica['PLANIRANI_DATUM_ZATVARANJA'] - tablica['DATUM_OTVARANJA']) < pd.Timedelta(days=0)].shape[0]))
    print('Koliko datuma ima DO == PDZ: ' + str(tablica[(tablica['PLANIRANI_DATUM_ZATVARANJA'] - tablica['DATUM_OTVARANJA']) == pd.Timedelta(days=0)].shape[0]))
    print('Koliko datuma ima DO < PDZ: ' + str(tablica[(tablica['PLANIRANI_DATUM_ZATVARANJA'] - tablica['DATUM_OTVARANJA']) > pd.Timedelta(days=0)].shape[0]))
    print('\n\n')
```

```python
analiziraj_datume(normalan_prvi_ugovor)
analiziraj_datume(prvi_ugovor_vracen_isti_dan)
analiziraj_datume(nenormalan_prvi_ugovor)
```

Oznake: 
- DO  ---> DATUM_OTVARANJA
- PDZ ---> PLANIRANI_DATUM_ZATVARANJA
