# Transformacija trening skupa

Ova bilježnica je namijenjena transformaciji trening skupa u skup koji ima isti 'format' kao i test skup. Uzet ćemo pickle file u kojem je originalni trening skup i od njega napraviti novi pickle file imena 'korigirani_trening_dataset_enc'

```python
import pandas as pd
import numpy as np
from script.feature_engineering import *
import functools
```

```python
los_train = pd.read_pickle('../data/training_dataset_enc.pkl')
```

```python
grupacija = los_train.groupby(['OZNAKA_PARTIJE', 'DATUM_OTVARANJA', 'PLANIRANI_DATUM_ZATVARANJA'])
```

```python
indeksi = np.array(
    [list(i) for i in grupacija.size().sort_values(ascending = False).index],
    dtype = object,
    order = 'C'
)
```

```python
indeksi
```

```python
# ovako izgleda jedna grupacija. Uzeli smo prvi indeks iz polja indeksi, a posto je sve sortirano po velicini
# tablica, ovo ispada kao najveca tablica
grupacija.get_group((4199321, pd.Timestamp('2009-01-21 00:00:00'),
        pd.Timestamp('2019-01-31 00:00:00')))
```

```python
indeksi.shape
```

```python
identiteta_nula = lambda stupac, grupa : grupa[stupac].iloc[0]

# definirano zbog PLANIRANI_DATUM_ZATVARANJA i VISINA_KAMATE
def uzmi_nenull(stupac, alt, grupa):
    i = grupa[stupac].first_valid_index()
    return alt if i is None else grupa.loc[i, stupac]

broj_razl = lambda stupac, grupa : pd.notnull(grupa[stupac].unique()).sum()
```

```python
import time
prije = time.time()

dobar_train = feat(
    grupacija,
    #tuple((tuple(indeks), grupacija.get_group(tuple(indeks))) for indeks in indeksi[:10000]),
    [
        ('KLIJENT_ID', functools.partial(identiteta_nula, 'KLIJENT_ID')),
        ('OZNAKA_PARTIJE', functools.partial(identiteta_nula, 'OZNAKA_PARTIJE')),
        ('DATUM_OTVARANJA', functools.partial(identiteta_nula, 'DATUM_OTVARANJA')),
        ('PLANIRANI_DATUM_ZATVARANJA', functools.partial(uzmi_nenull, 'PLANIRANI_DATUM_ZATVARANJA', pd.NaT)),
        ('UGOVORENI_IZNOS', functools.partial(identiteta_nula, 'UGOVORENI_IZNOS')),
        ('VALUTA', functools.partial(identiteta_nula, 'VALUTA')),
        ('VRSTA_KLIJENTA', functools.partial(identiteta_nula, 'VRSTA_KLIJENTA')),
        ('PROIZVOD', functools.partial(identiteta_nula, 'PROIZVOD')),
        ('VRSTA_PROIZVODA', functools.partial(identiteta_nula, 'VRSTA_PROIZVODA')),
        ('VISINA_KAMATE', functools.partial(uzmi_nenull, 'VISINA_KAMATE', np.nan)),
        ('TIP_KAMATE', functools.partial(identiteta_nula, 'TIP_KAMATE')),
        ('STAROST', functools.partial(identiteta_nula, 'STAROST')),
        (
            'PRIJEVREMENI_RASKID',
            lambda grupa : (grupa.DATUM_ZATVARANJA + pd.Timedelta(10, 'D') < grupa.PLANIRANI_DATUM_ZATVARANJA).any()
        )
    ]
)

print('Vrijeme: ' + str(time.time()-prije))
```

```python
dobar_train.to_pickle('../data/dobar_trening_set.pkl')
```
