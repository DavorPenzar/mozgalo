```python
import pandas as pd
```

## **Učitavanje tablice**

```python
df = pd.read_excel('ekonomski-pokazatelji/ekonomski_indikatori.xlsx', sheet_name = 'HRVIND')

df.head()
```

```python
df.tail(10)
```

## **Uređivanje tablice**

```python
df = df.drop(['Unnamed: 0'], axis=1)
df = df.drop([0, 1, 2])
df.drop(df.tail(9).index, inplace=True) ## odbaci zadnjih 9 redaka
```

```python
df
```

```python
godine = df.iloc[0]
df.columns = godine ## pretvaranje prvog reda u column header
df = df[3:] ## odbacivanje tog prvog reda, površine RH i evidencije njenog stanovništva
df.head()
```

```python
df.columns = df.columns.fillna('Info') ## zamijeni nan vrijednost sa 'Info'

## uredivanje header-a
col_list = list(df.columns)
col_list[1] = 2000
col_list[2] = 2001
col_list[len(col_list) - 1] = 2018
for i in range(3, len(col_list) - 1):
    col_list[i] = int(col_list[i])
df.columns = col_list
df.head(16)
```

```python
df = df.drop([21]) ## odbacivanje retka koji u sebi ima samo vrijednosti NaN
df.index = range(len(df)) ## enumeracija redaka od 0 pa nadalje 
df
```

```python
## spremi u pickle
df.to_pickle('ekonomski-pokazatelji/ekonomski_indikatori.pkl')
```
